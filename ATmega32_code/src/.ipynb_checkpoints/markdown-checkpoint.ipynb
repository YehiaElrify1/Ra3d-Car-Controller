المؤسسة الأكاديمية: المعهد العالي للهندسة والتكنولوجيا بكفر الشيخ

التردد الرئيسي للمنظومة: 16 MHz External Crystal Oscillator

1. مقدمة علمية في الأنظمة المدمجة (Introduction to Embedded Systems)

الأنظمة المدمجة (Embedded Systems): هي أنظمة حاسوبية متخصصة ومصممة لأداء وظائف تحكم محددة ومكرسة داخل أنظمة ميكانيكية أو كهربائية أكبر، وتعمل غالباً تحت قيود الزمن الحقيقي (Real-Time Constraints).

في هذا المشروع الأكاديمي، تم استخدام ميكروكنترولر ATmega32 المبني على معمارية 8-bit AVR RISC المصنعة من شركة Microchip/Atmel. لضمان استقرار العمليات البرمجية والتزامن الدقيق للبيانات التسلسلية وحسابات تعديل عرض النبضة (PWM)، تم تزويد المتحكم بكريستالة خارجية بتردد 16 MHz.

هذا التردد العالي يمنح المعالج القدرة على تنفيذ $16$ مليون تعليمة في الثانية الواحدة (16 MIPS)، وهو ما يضمن استجابة فورية للأعطال وأنظمة الأمان اللحظية دون أي تأخير ميكروثاني ملحوظ.

2. التشريح العتادي للوحدات الطرفية وسجلاتها (Hardware Peripherals & Registers)

تعتمد برمجة الأنظمة المدمجة الاحترافية على التحكم المباشر في السجلات الفيزيائية (Hardware Registers) الخاصة بالملحقات الطرفية (Peripherals). سنقوم هنا بتفكيك ووصف كل بت (Bit) وسجل تم استخدامه في الكود الخاص بك.

أ. وحدة الإدخال والإخراج الرقمية (GPIO)

هي الواجهة الأساسية التي تمكن الميكروكنترولر من التعامل مع المكونات الرقمية الخارجية عن طريق قراءة مستويات الجهد المنطقي أو قيادتها. يتم التحكم بالـ GPIO عبر ثلاثة سجلات رئيسية لكل منفذ (Port):

DDRx (Data Direction Register): يحدد اتجاه الدبابيس (إدخال $0$ أم إخراج $1$).

PORTx (Data Register): يحدد حالة الدبوس المخرج (مرتفع High أم منخفض Low)، أو يفعل المقاومة الرافعة الداخلية (Internal Pull-up) في حالة الإدخال.

PINx (Input Pins Address): سجل مخصص لقراءة مستويات الجهد الفيزيائي الخارجي الواقع على الدبوس في تلك اللحظة.

تطبيق ذلك في الكود الخاص بك:

DDRA |= 0x0F; تم تهيئة الدبابيس من PA0 إلى PA3 كمخارج رقمية لتغذية اتجاهات درايفر المحركات L298N.

DDRA |= (1 << PA4); تم تهيئة الدبوس PA4 كمخرج لإرسال نبضة التشغيل (Trig) لحساس الـ Ultrasonic.

DDRA &= ~(1 << PA5); تم تهيئة الدبوس PA5 كمدخل لاستقبال إشارة الصدى (Echo).

ب. وحدة التواصل التسلسلي غير المتزامن (UART)

بروتوكول اتصال تسلسلي يعتمد على التوقيت المتفق عليه مسبقاً (Baud Rate) دون الحاجة لخط ساعة مشترك (Clock Line). تم ضبطه كالتالي:

UCSRB (Control and Status Register B):
تفعيل التمكين للإرسال والاستقبال عبر الأوامر البرمجية:

UCSRB = (1 << RXEN) | (1 << TXEN);


البت RXEN (Receiver Enable) يفعل استقبال البيانات عتادياً، والبت TXEN (Transmitter Enable) يفعل إرسال البيانات عتادياً عبر منفذ الـ UART.

UCSRC (Control and Status Register C):

UCSRC = (1 << URSEL) | (1 << UCSZ1) | (1 << UCSZ0);


البت URSEL (Register Select) يجب أن يكتب بـ $1$ للتمكن من الكتابة في سجل UCSRC بدلاً من سجل UBRRH لكونهما يتشاركان نفس العنوان الفيزيائي داخل المعالج. البتان UCSZ1 و UCSZ0 تم تفعيلهما معاً لضبط حجم بيانات الإطار (Character Size) ليكون بعرض 8-bit.

ج. المحول الرقمي التماثلي (ADC)

يقوم بتحويل الجهد المستمر والتماثلي إلى قيمة رقمية ممثلة بخطوات متقطعة بناءً على دقة التحليل (Resolution) الخاصة بالمحول.

ADMUX (Multiplexer Selection Register):

ADMUX = (1 << REFS0);


البت REFS0 يحدد الجهد المرجعي للتحويل ليكون مساوياً لجهد التغذية الرئيسي AVCC ($5\text{ V}$) مع وضع مكثف خارجي عند طرف AREF لتقليل تداخل الإشارات والضوضاء الكهربائية.

ADCSRA (Control and Status Register A):

ADCSRA = (1 << ADEN) | (1 << ADPS2) | (1 << ADPS1) | (1 << ADPS0);


البت ADEN (ADC Enable) يفعل دائرة المحول عتادياً. البتات الثلاثة ADPS2, ADPS1, ADPS0 تحدد معامل تقسيم تردد الساعة (Prescaler) ليكون مساوياً لـ 64.

د. المؤقت الأول وتوليد الـ PWM (Timer 1)

تقنية تعديل عرض النبضة (Pulse Width Modulation - PWM) تتيح للمتحكم الرقمي إخراج جهد متوسط متغير عبر التحكم في زمن التشغيل (Duty Cycle) للإشارة المربعة. تم برمجة المؤقت الأول (Timer1) وهو مؤقت بدقة 16-bit للتحكم بمحركات التيار المستمر:

TCCR1A (Timer/Counter 1 Control Register A):

TCCR1A = (1 << COM1A1) | (1 << COM1B1) | (1 << WGM10);


البتان COM1A1 و COM1B1 يفعلان نظام المقارنة غير العاكس (Non-Inverting Mode) لقناتي المقارنة Channel A و Channel B. هذا يعني أن الجهد على دبوس المخرج سيبدأ بحالة مرتفعة (High) ويتحول لمنخفض (Low) بمجرد أن تتطابق قيمة المؤقت مع قيمة السجل المقارن OCR1x. البت WGM10 مع البت WGM12 في السجل الثاني يحددان تشغيل النظام في وضع Fast PWM 8-bit (Mode 5).

3. الاشتقاق والتحليل الرياضي والهندسي (Mathematical Derivations)

أ. اشتقاق سجل نقل البيانات (UART Baud Rate Derivation)

في وضع الاتصال غير المتزامن العادي (Normal Asynchronous Mode)، تحسب قيمة السجل المسؤول عن التوقيت UBRR عبر الصيغة الرياضية التالية:

$$UBRR = \frac{F_{CPU}}{16 \times \text{Baud Rate}} - 1$$

حيث أن:

$F_{CPU} = 16,000,000 \text{ Hz}$ (تردد الكريستالة الخارجية المتصلة بالـ ATmega32).

$\text{Baud Rate} = 9600 \text{ bps}$ (معدل البث المتفق عليه مع موديول HC-05).

بالتعويض المباشر في القانون:

$$UBRR = \frac{16,000,000}{16 \times 9600} - 1 = \frac{16,000,000}{153,600} - 1$$

$$UBRR = 104.166 - 1 = 103.166 \approx 103$$

لحساب نسبة الخطأ المئوية الفعلية في نقل البيانات (Baud Rate Error Rate):

$$\text{Baud Rate}_{\text{Actual}} = \frac{16,000,000}{16 \times (103 + 1)} \approx 9615.38 \text{ bps}$$

$$\text{Error \%} = \left| \frac{9615.38 - 9600}{9600} \right| \times 100\% \approx 0.16\%$$

هذه النسبة الضئيلة جداً (أقل من $1\%$) تعتبر ممتازة وتضمن استقرارية كاملة للاتصال التسلسلي دون حدوث أي فقدان للبيانات أو تداخل إشارات (Framing Errors).

ب. اشتقاق معادلة تحويل قراءة الحرارة (ADC Temperature Derivation)

المحول الرقمي في الـ ATmega32 يعتمد على دقة 10-bit، أي أنه يقسم إشارة المدخل التماثلي إلى $2^{10} = 1024$ مستوى رقمي خطي تتراوح من $0$ إلى $1023$.

حجم خطوة الجهد الفردية (Voltage Step Size):

$$\text{Step Size} = \frac{V_{REF}}{1024} = \frac{5 \text{ V}}{1024} = 0.0048828 \text{ V} \approx 4.88 \text{ mV}$$

بما أن حساس الحرارة التماثلي الشهير LM35 يخرج جهداً خطياً بمعدل $10 \text{ mV}$ لكل درجة مئوية واحدة ($10 \text{ mV}/^\circ\text{C}$)، فإن جهد مخرج الحساس يعبر عنه بـ:

$$V_{out} = T \times 10 \text{ mV} = T \times 0.01 \text{ V}$$

حيث تمثل $T$ درجة الحرارة الفعلية. القيمة الرقمية الناتجة من قراءة السجل (ADC_Value) هي:

$$\text{ADC\_Value} = \frac{V_{out}}{V_{REF}} \times 1024 = \frac{T \times 0.01}{5} \times 1024$$

بإعادة الترتيب الرياضي لحل المعادلة بالنسبة لدرجة الحرارة $T$:

$$T = \text{ADC\_Value} \times \frac{5}{1024 \times 0.01} = \text{ADC\_Value} \times \frac{500}{1024}$$

وهذه هي المعادلة المطبقة برمجياً بدقة فائقة لتفادي الكسور العشرية الثقيلة على المعالجات الـ 8-bit:

temp_c = (uint32_t)ADC_Read(6) * 500 / 1024;


تحليل هندسي متقدم لمعدل أخذ العينات (ADC Prescaler Trade-off):
تردد ساعة الـ ADC المثالي لتحقيق أعلى دقة (10-bit Accuracy) يجب أن يتراوح بين $50 \text{ kHz}$ و $200 \text{ kHz}$. في الكود الحالي، قمت بضبط الـ Prescaler على $64$:

$$f_{ADC} = \frac{F_{CPU}}{64} = \frac{16,000,000}{64} = 250 \text{ kHz}$$

هذا التردد ($250 \text{ kHz}$) يقع أعلى بقليل من المدى الموصى به للمثالية ($200 \text{ kHz}$). هذا الاختيار يمثل مقايضة هندسية (Trade-off) حيث يمنحك سرعة تحويل أعلى ومعدل استجابة أسرع لقراءة درجة الحرارة والتحكم مع فقدان ضئيل للغاية يكاد يكون غير محسوس في الدقة الرقمية، وهو اختيار مثالي لتطبيقات الـ Dynamic Telemetry والتحكم اللحظي.

ج. اشتقاق تردد إشارة الـ PWM لسرعة المحركات (PWM Frequency)

بناءً على اختيار وضع الـ Fast PWM 8-bit بمقسم تردد $64$، يتم تحديد تردد الإشارة المربعة عبر المعادلة:

$$f_{PWM} = \frac{F_{CPU}}{N \times 256}$$

حيث تمثل $N$ مقسم التردد الـ Prescaler. بالتعويض:

$$f_{PWM} = \frac{16,000,000}{64 \times 256} = \frac{16,000,000}{16,384} \approx 976.56 \text{ Hz}$$

هذا التردد (حوالي $1 \text{ kHz}$) يعتبر مثالياً جداً لمحركات التيار المستمر (DC Motors). الترددات الأقل من ذلك قد تؤدي إلى حدوث اهتزاز ميكانيكي مسموع (صوت زنين مزعج) في ملفات المحرك، في حين أن الترددات الأعلى بكثير قد تتسبب في زيادة خسائر التبديل الكهربائي (Switching Losses) داخل ترانزستورات درايفر L298N وارتفاع درجة حرارته بشكل خطير.

4. شرح الهندسة البرمجية والأمان (Software Engineering & Safety)

أ. الحماية البرمجية المتقدمة لسلامة المستشعرات (Bitwise Masking)

في الأنظمة البرمجية المدمجة الاحترافية، يمنع تماماً الكتابة المباشرة على كامل السجل الرقمي دون الأخذ بعين الاعتبار الوظائف الأخرى المتصلة بنفس السجل. يظهر ذلك بوضوح في كودك في معالجة أوامر الحركة:

PORTA = (PORTA & 0xF0) | 0b00000101;


التشريح المنطقي للعملية:

القناع الثنائي (Masking): عملية الـ Bitwise AND بالقناع 0xF0 (الذي يقابل بالباينري 11110000) تقوم بالحفاظ على حالة الدبابيس الأربعة العلوية (PA4-PA7) وتصفير الدبابيس الأربعة السفلية (PA0-PA3) تماماً.

الحفاظ على المستشعرات: الدبوسان PA4 و PA5 مخصصان لحساس الـ Ultrasonic. بفضل هذا القناع، لا يمكن لأي أمر حركة قادم من البلوتوث أن يغير حالة دبابيس الحساس أو يشوش على قراءة المسافات بالخطأ.

دمج حالة الحركة: نقوم بعمل Bitwise OR مع قيمة الحركة الجديدة (مثلاً 0b00000101 للحركة للأمام) لتتحرك السيارة فوراً دون المساس ببقية أطراف الحساسات في النظام.

ب. تجنب مشاكل تعليق النظام (Non-Blocking Design)

تعتمد دالة الاستقبال على فحص علم وصول البيانات (Polling the RXC Flag) بشكل لحظي ودون تعليق:

if (UCSRA & (1 << RXC)) return UDR;
return 0;


تجنبنا تماماً استخدام الحلقات التكرارية اللانهائية للانتظار مثل (while (!(UCSRA & (1 << RXC)));). هذا التصميم يسمى الـ Non-Blocking Design، وهو ضروري لضمان عدم توقف المعالج عن قراءة المسافات وتأمين السيارة من الاصطدام أثناء انتظار وصول أمر من الهاتف عبر البلوتوث.

6. توصيات هندسية متقدمة للتطوير المستقبلي

Input Capture Unit (ICU): بدلاً من قياس زمن نبضة الصدى (Echo) الخاصة بحساس المسافة عبر الـ Polling البرمجي المعطل للمعالجة الزمنية، يُفضل مستقبلاً استخدام وحدة الـ ICU المتصلة بـ Timer1 عتادياً لقياس الزمن بالنانوثانية بمجرد تغير مستوى الجهد دون أي تدخل برمي يعطل المعالج الرئيسي.

Noise Decoupling & Filtering: لمكافحة الضوضاء الناتجة عن المحركات الحثية (DC Motors)، يوصى بفصل تام للتغذية ووضع مكثفات تنعيم سيراميكية بسعة $0.1\mu\text{F}$ بالتوازي مع مكثفات إليكتروليتية بسعة $10\mu\text{F}$ بالقرب من دبابيس طاقة الميكروكنترولر مباشرة لمنع الـ Spikes والـ Reset العشوائي.
